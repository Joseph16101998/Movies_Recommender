###### creating helper function that will help us obtain the above results

```python
#The will help us convert strings of dictionary into list of dictionary
import ast
```

```python
#before lets take backup so that we don't need to start over
moviesbck=movies_df.copy()
```

```python
def fetch_genre(obj):
    list_genre=[]
    for i in ast.literal_eval(obj):
        list_genre.append(i['name'])
    return list_genre

def fetch_keywords(obj):
    list_kw=[]
    for i in ast.literal_eval(obj):
        list_kw.append(i['name'])
    return list_kw

def fetch_cast(obj):
    list_cast=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter !=3:
            list_cast.append(i['name'])
            counter+=1
        else:
            break
    return list_cast

def fetch_director(obj):
    list_director=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list_director.append(i['name'])
            break
    return list_director
```

```python
movies_df['genres']=movies_df['genres'].apply(fetch_genre)
```

```python
movies_df['keywords']=movies_df['keywords'].apply(fetch_keywords)
```

```python
movies_df['cast']=movies_df['cast'].apply(fetch_cast)
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_director)
```

###### Now we will create a vector of all the keywords from our 'tags' columns
- for that we would require 'from sklearn.feature_extraction.text import CountVectorizer'

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#instantiating the CountVectorizer class
cv=CountVectorizer(max_features=5000,stop_words='english')

```

```python
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
# to get all the names 
cv.get_feature_names()
```

###### We can see that there are few words repeated like 'actions','action' similarly our dataset can have multiple amount of repeatition.
- This can be resolve using 'from nltk.stem.porter import PorterStemmer'

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
#Instantiating PortStemmer class
ps=PorterStemmer()
#we will apply stem on our 'tags' column
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
#we have now eliminated the repeated words
cv.get_feature_names()
```

```python
#lets take backup of the transform dataset
moviesbcktr=movies.copy()
moviesbcktr.head()
```

###### Now we have converted our 'tags' columns to a vector for comparing each vector with eachother
- To achieve this we will be using library that will do the work for us 'from sklearn.metrics.pairwise import cosine_similarity'

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
similarity.shape
```

```python
#main recommendation logic
def recommend(movie):
    movie_idx=movies[movies['title']==movie].index[0]
    distances=similarity[movie_idx]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
      
```

```python
recommend('Gandhi')
```
